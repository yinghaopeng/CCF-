```python
import matplotlib.pyplot as plt
import seaborn as sns
import gc  # python垃圾回收机制
import re
import pandas as pd
import lightgbm as lgb
import numpy as np
from sklearn.metrics import roc_auc_score, precision_recall_curve, roc_curve, average_precision_score
from sklearn.model_selection import KFold, StratifiedKFold
from lightgbm.sklearn import LGBMClassifier
from dateutil.relativedelta import relativedelta
import warnings
warnings.filterwarnings("ignore")
pd.set_option('max_columns', None)  # 显示所有的列数
pd.set_option('max_rows', 200)  # 显示最多200行
pd.set_option('float_format', lambda x: '%.3f' % x)  # 小数位数3位

train_data = pd.read_csv('train_public.csv')
submit_example = pd.read_csv('submit_example.csv')
test_public = pd.read_csv('test_public.csv')
train_inte = pd.read_csv('train_internet.csv')
```

```python
def train_model(data, test, y, fold):
    oof_preds = np.zeros(data.shape[0])
    sub_preds = np.zeros(test.shape[0])
    feature_importance_df = pd.DataFrame()
    features = [f for f in data.columns if f not in ['loan_id', 'user_id', 'isDefault']]
    for n_fold, (train_index, test_index) in enumerate(fold.split(data)):
       tr_X, tr_Y = data[features].iloc[train_index], y.iloc[train_index]
       val_X, val_Y = data[features].iloc[test_index], y.iloc[test_index]
       LGBM_model = LGBMClassifier(
        n_estimators = 4000, 
        learning_rate=0.08, 
        num_leaves=2**5, 
        colsample_bytree=.65, 
        subsample=.9, 
        max_depth=5, 
        reg_alpha=.3, 
        reg_lambda=.3, 
        min_split_gain=.01, 
        min_child_weight=2, 
        silent=-1, 
        verbose=-1
        )

       LGBM_model.fit(tr_X, tr_Y, 
                      eval_set = [(tr_X, tr_Y), (val_X, val_Y)], 
                      eval_metric = 'auc', verbose=100, early_stopping_rounds=40
         )
       oof_preds[test_index] = LGBM_model.predict_proba(val_X, num_iteration=LGBM_model.best_iteration_)[:, 1]
       sub_preds += LGBM_model.predict_proba(test[features], num_iteration=LGBM_model.best_iteration_)[:, 1] / fold.n_splits
      
       fold_importance_df = pd.DataFrame()
       fold_importance_df["feature"] = features
       fold_importance_df["importance"] = LGBM_model.feature_importances_
       fold_importance_df["fold"] = n_fold + 1
       feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)

       print('Fold %2d AUC : %.6f' % (n_fold + 1, roc_auc_score(val_Y, oof_preds[test_index])))
       del LGBM_model, tr_X, tr_Y, val_X, val_Y
       gc.collect()
    print('Full AUC score %.6f' % roc_auc_score(y, oof_preds))

    test['isDefault'] = sub_preds
    
    return oof_preds, test[['loan_id', 'isDefault']], feature_importance_df

# 将work_year列转化成数字
def work_year_transform(x):
    pattern = '\d+'
    if(str(x) == 'nan'):  #缺失值用-1代替
        return -1
    x = x.replace('< 1', '0')
    return int(re.search(pattern,  x).group())
# 将earlies_credit_mon列转化成datetime格式
def earlies_credit_mon_transform(x):
    pattern = '\d+-'
    res = re.search(pattern, x)
    if(res is None):
        return '1-'+x
    return x + '-01'

def display_importances(feature_importance_df_):
    # Plot feature importances
    cols = feature_importance_df_[["feature", "importance"]].groupby("feature").mean().sort_values(
        by="importance", ascending=False)[:50].index
    
    best_features = feature_importance_df_.loc[feature_importance_df_.feature.isin(cols)]
    
    plt.figure(figsize=(8,10))
    sns.barplot(x="importance", y="feature", 
                data=best_features.sort_values(by="importance", ascending=False))
    plt.title('LightGBM Features (avg over folds)')
    plt.tight_layout()
    plt.savefig('lgbm_importances.png')
```

```python
# public数据处理
class_grade = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,}
timeMax = pd.to_datetime('2021-12-31', format='%Y-%m-%d')
train_data['work_year'] = train_data['work_year'].map(work_year_transform)
test_public['work_year'] = test_public['work_year'].map(work_year_transform)
train_data['class'] = train_data['class'].map(class_grade)
test_public['class'] = test_public['class'].map(class_grade)
train_data['earlies_credit_mon'] = pd.to_datetime(train_data['earlies_credit_mon'].map(earlies_credit_mon_transform))
test_public['earlies_credit_mon'] = pd.to_datetime(test_public['earlies_credit_mon'].map(earlies_credit_mon_transform))
# 将超过2021年的数据进行处理
train_data.loc[train_data['earlies_credit_mon'] > timeMax, 'earlies_credit_mon'] = train_data.loc[train_data['earlies_credit_mon'] > timeMax, 'earlies_credit_mon'] + pd.offsets.DateOffset(years = -100)
test_public.loc[test_public['earlies_credit_mon'] > timeMax, 'earlies_credit_mon'] = test_public.loc[test_public['earlies_credit_mon'] > timeMax, 'earlies_credit_mon'] + pd.offsets.DateOffset(years = -100)
# 将issue_date转化成datetime格式
train_data['issue_date'] = pd.to_datetime(train_data['issue_date'])
test_public['issue_date'] = pd.to_datetime(test_public['issue_date'])
```

```python
# 对Internet数据处理
train_inte['work_year'] = train_inte['work_year'].map(work_year_transform)
train_inte['class'] = train_inte['class'].map(class_grade)
train_inte['earlies_credit_mon'] = pd.to_datetime(train_inte['earlies_credit_mon'])
train_inte['issue_date'] = pd.to_datetime(train_inte['issue_date'])

# 对issue_date进行特征生成
train_data['issue_date_month'] = train_data['issue_date'].dt.month
train_data['issue_date_dayofweek'] = train_data['issue_date'].dt.dayofweek
test_public['issue_date_month'] = test_public['issue_date'].dt.month
test_public['issue_date_dayofweek'] = test_public['issue_date'].dt.dayofweek
# Internet数据
train_inte['issue_date_month'] = train_inte['issue_date'].dt.month
train_inte['issue_date_dayofweek'] = train_inte['issue_date'].dt.dayofweek

# 对earlies_credit_mon进行特征生成
train_data['earlies_credit_mon_Month'] = train_data['earlies_credit_mon'].dt.month
train_data['earlies_credit_mon_Year'] = train_data['earlies_credit_mon'].dt.year
test_public['earlies_credit_mon_Month'] = test_public['earlies_credit_mon'].dt.month
test_public['earlies_credit_mon_Year'] = test_public['earlies_credit_mon'].dt.year
# Internet数据
train_inte['earliesCreditMon'] = train_inte['earlies_credit_mon'].dt.month
train_inte['earliesCreditYear'] = train_inte['earlies_credit_mon'].dt.year
```

```python
# 对'employer_type',和'industry'进行编码处理
no_numerical_fea= ['employer_type', 'industry']
from sklearn.preprocessing import LabelEncoder
for col in no_numerical_fea:
    LE = LabelEncoder().fit(train_data[col])
    train_data[col] = LE.transform(train_data[col])
    test_public[col] = LE.transform(test_public[col])
    # Internet数据
    train_inte[col] = LE.transform(train_inte[col])

# 前面对'issue_date'和'earlies_credit_mon'处理后可以删除原本列
dele_fea= ['issue_date', 'earlies_credit_mon']
train_data = train_data.drop(dele_fea, axis = 1)
test_public = test_public.drop(dele_fea, axis = 1)
##internet处理
train_inte = train_inte.drop(dele_fea, axis = 1)
```

```python
train_data_cols = set(train_data.columns)
# public和Internet相同的列
same_cols = list(train_data_cols.intersection(set(train_inte.columns)))
train_inte_same_cols = train_inte[same_cols].copy()

train_inte_add_cols = list(train_data_cols.difference(set(same_cols)))
for col in train_inte_add_cols:
    train_inte_same_cols[col] = np.nan
```

```python
y = train_data['isDefault']
folds = KFold(n_splits=5, shuffle=True, random_state=546789)
oof_preds, IntePre, importances = train_model(train_data, train_inte_same_cols, y, folds)
IntePre['is_Default'] = train_inte['is_default']
roc_auc_score(IntePre['is_Default'], IntePre['isDefault'])
# 选择阈值0.05，从internet表中提取预测小于该概率的样本，并对不同来源的样本赋予来源值
InteId = IntePre.loc[IntePre['isDefault'] < 0.5, 'loan_id'].tolist()

train_data['dataSourse'] = 1
test_public['dataSourse'] = 1
train_inte_same_cols['dataSourse'] = 0
train_inte_same_cols['isDefault'] = train_inte['is_default']
use_te = train_inte_same_cols[train_inte_same_cols.loan_id.isin(InteId)].copy()
data = pd.concat([train_data, test_public, use_te]).reset_index(drop = True)

plt.figure(figsize=(16,6))
plt.title("Distribution of Default values IntePre")
sns.distplot(IntePre['isDefault'],color="black", kde=True,bins=120, label='train_data')
plt.legend();plt.show()
train = data[data['isDefault'].notna()]
test  = data[data['isDefault'].isna()]

del data
del train_data,test_public
y = train['isDefault']
folds = KFold(n_splits = 5, shuffle = True, random_state = 546789)
oof_preds, test_preds, importances = train_model(train, test, y, folds)
test_preds.rename({'loan_id': 'id'}, axis=1)[['id', 'isDefault']].to_csv('result.csv', index=False)
```
